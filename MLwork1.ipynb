```python
def input_cabin(cols):
    Cabin = cols[0]
    if pd.isnull(Cabin):
        return 0
    else:
        return 1
    
```

```python
train['Cabin'] = train[['Cabin']].apply(input_cabin, axis=1)
```

```python
train.head(10)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=True,cmap='viridis')
plt.show()
```

```python
train.isnull().sum()
```

```python

train.drop('Age',axis=1,inplace=True)
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
train.head()
```

```python
train.dropna(inplace=True)
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
train.info()
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
sex
```

```python
embark = pd.get_dummies(train['Embarked'],drop_first=True)
embark
```

```python

train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
train = pd.concat([train,sex,embark],axis=1)
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split

# Correctly split the data
x_train, x_test, y_train, y_test = train_test_split(
    train.drop('Survived', axis=1),
    train['Survived'],
    test_size=0.20,
    random_state=101
)
```

# Training And Predicting

```python

from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel
```

```python
logmodel.fit(x_train,y_train)
```

```python
prediction = logmodel.predict(x_test)
prediction
```

```python
y_test
```

```python
# we can check precision, recall, f1-score using classification report
from sklearn.metrics import auc, accuracy_score,classification_report,log_loss,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
```

```python
print(sns.heatmap(confusion_matrix(prediction,y_test),annot=True))
```
